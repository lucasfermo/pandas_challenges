```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from mpl_toolkits import mplot3d
from matplotlib import dates as date
import seaborn as sns; sns.set() 
from scipy import stats
from datetime import datetime
plt.style.use("fivethirtyeight")
```

```python
#Import dataset
data={"first_name":['John','Molly',"Tina",'Jake','Amy'],
     'last_name':['Miller','Jacobson','Ali','Milner','Cooze'],
     'age':[42,52,36,24,73],
     'female':[0,1,1,0,1],
     'preTestScore':[4,24,31,2,3],
         'postTestScore':[25,94,57,62,70]}
scores=pd.DataFrame(data,columns=['first_name','last_name','age','female',
                                 'preTestScore','postTestScore'])
```

```python
#Plot scatter of scores
plt.scatter(scores.preTestScore,scores.postTestScore,s=scores.age)
plt.xlabel("Pre-Test Scores")
plt.ylabel("Post-Test Scores")
plt.title('Test comparison')
```

```python
#Increase size by 4.5
plt.scatter(scores.preTestScore,scores.postTestScore,s=scores.age*4.5)
plt.xlabel("Pre-Test Scores")
plt.ylabel("Post-Test Scores")
plt.title('Test comparison')
```
