```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from mpl_toolkits import mplot3d
from matplotlib import dates as date
import seaborn as sns; sns.set() 
from scipy import stats
from datetime import datetime
plt.style.use("fivethirtyeight")
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Plot histogram of bill total
tbill=sns.distplot(tips.total_bill)
tbill.set(xlabel="Total Bill",ylabel="Frequency",title='Total Bill')
```

```python
#Plot scatter between total_bill and tip
tbill=sns.jointplot('total_bill','tip',data=tips)
```

```python
#Plot relationship between total bill,size and tips
sns.pairplot(tips)
```

```python
#Plot relationship between day and total bill
sns.stripplot('day','total_bill',data=tips,jitter=True)
```

```python
#Plot relationship betewen total_bill and day, differing sex by dots
ax=sns.stripplot('tip','day',data=tips,hue='sex',jitter=True)
ax.set(xlabel="Tip amount",ylabel='Day')
```

```python
#Plot boxplot comparing total bill to day, differentiating the time
sns.boxplot('day','total_bill',hue='time',data=tips)
```

```python
#Plot histograms side-by-side for lunch and dinner
g=sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

```python
#Plot two scatterplots side-by side, one for male and female, presenting
#total bill value and tip relationship
g=sns.FacetGrid(tips,col='sex',hue='smoker')
g.map(plt.scatter,'total_bill','tip',alpha=0.5)
```
