```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from pandas_datareader.data import DataReader
from mpl_toolkits import mplot3d
from matplotlib import dates as date
import seaborn as sns; sns.set() 
from scipy import stats
from datetime import datetime
plt.style.use("fivethirtyeight")
```

```python
#Import apple stock prices
apple=DataReader('AAPL',start=2012,end=2017,data_source='google')
```

```python
#Make first entry from the oldest date
apple=apple[::-1]
```

```python
#Get the last business day of each month
apple_last=apple.resample("BM").mean()
```

```python
#Difference between first and last date
apple.index[0]-apple.index[-1]
```

```python
#How many months of data do we have
len(apple.resample('M').mean())
```

```python
#Plot the close value, set the figure to 13.5 x 9 inches
fig,ax=plt.subplots()
ax.plot(apple.Close)
ax.set_title('Apple Stock')
fig.set_size_inches(13.5,9)
```
