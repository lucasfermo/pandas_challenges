```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from mpl_toolkits import mplot3d
from matplotlib import dates as date
import seaborn as sns; sns.set() 
from scipy import stats
plt.style.use("fivethirtyeight")
```

```python
#Import dataset
baby_names=pd.read_csv("data/US_Baby_Names_right.csv")
```

```python
#Delete the column Unnamed:0 and Id
baby_names=baby_names.loc[:,'Name':'Count']
```

```python
#Is there more males or females in dataset
baby_names['Gender'].value_counts()
```

```python
#Group the dataset by name
names=baby_names.groupby("Name").sum()
del names['Year']
```

```python
#Name with most occurences
names[names.Count==names.Count.max()]
```

```python
#Name With least occurence
names[names.Count==names.Count.min()]

#Name with median occurence
names[names.Count==names.Count.median()]
```

```python
#Standard Deviation of names
names.Count.std()
```

```python
#Get a summary with the mean, min,max and std quartiles
names.describe()
```
