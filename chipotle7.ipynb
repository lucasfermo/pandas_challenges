```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from mpl_toolkits import mplot3d
from matplotlib import dates as date
import seaborn as sns; sns.set() 
from scipy import stats
from datetime import datetime
plt.style.use("fivethirtyeight")
```

```python
url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo=pd.read_csv(url,sep="\t")
```

```python
#Plot histogram of most popular items
y=chipo.groupby("item_name").size().sort_values(ascending=False)[:5]
y.plot(kind='bar')
```

```python
#Create a scatter plot of price on x and numbers of orders on y
by_price=chipo.groupby("item_price").size()
price=by_price.index.astype(str)
price=[i for i in price]
price=list(map(lambda x: float(x[1:]),price))
plt.scatter(price,by_price)
```
