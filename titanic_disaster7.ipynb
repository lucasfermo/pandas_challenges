```python
titanic=pd.read_csv("data/train.csv")
```

```python
titanic.head()
```

```python
titanic=titanic.set_index('PassengerId')
```

```python
#Create pie chart presenting male/female proportion
males=(titanic['Sex']=='male').sum()
females=(titanic['Sex']=='female').sum()

proportions=[males,females]

plt.pie([males,females],labels=['Males','Females'],explode=(0.1,0),autopct="%1.1f%%")

plt.axis('equal')
plt.title('Sex Proportion')
```

```python
#Create a scatterplot with the Fare payed and the age,differ plot color by gender
ax=sns.lmplot('Age','Fare',data=titanic,fit_reg=False,hue='Sex')
```
